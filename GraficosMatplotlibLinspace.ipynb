```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,num=10)
y = np.linspace(0,10,num=10)

z = np.linspace(0,10,num=10)
w = np.linspace(1,11,num=10)

plt.title("Retas Paralelas")
plt.plot(x,y)
plt.plot(z,w)
```

```python
a = np.linspace(0,10,num=10)
b = np.linspace(10,0,num=10)

c = np.linspace(0,10,num=10)
d = np.linspace(1,21,num=10)

plt.title("Retas Concorrentes")
plt.plot(a,b)
plt.plot(c,d)
```
